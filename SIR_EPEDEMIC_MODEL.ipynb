```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Total population, N.
N = 3000000
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 100, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma = 0.167, 1./15
# A grid of time points (in days)
t = np.linspace(0, 160, 160)

# The SIR model differential equations.
def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

# Initial conditions vector
y0 = S0, I0, R0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T

# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, facecolor='#dddddd', axisbelow=False)
ax.plot(t, S/1000, 'b', alpha=0.8, lw=5, label='Susceptible')
ax.plot(t, I/1000, 'r', alpha=0.8, lw=5, label='Infected')
ax.plot(t, R/1000, 'g', alpha=0.8, lw=5, label='Recovered with immunity')
ax.set_xlabel('Time /days')
ax.set_ylabel('Population')
#ax.set_ylim(0,1.2)
#ax.yaxis.set_tick_params(length=10)
#ax.xaxis.set_tick_params(length=10)
#ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
#for spine in ('top', 'right', 'bottom', 'left'):
    #ax.spines[spine].set_visible(True)
plt.show()
```
